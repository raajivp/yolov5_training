```python
import os

dataset_path = "/content/Data Set"  # Replace with the path to your dataset from Task 1
all_image_paths = [os.path.join(root, fname) for root, _, filenames in os.walk(dataset_path) for fname in filenames if fname.endswith('.jpg')]
```

```python
from transformers import AutoImageProcessor

image_processor = AutoImageProcessor.from_pretrained("facebook/dinov2-base")
```

```python
import torch
```

```python
from transformers import Dinov2Model

model = Dinov2Model.from_pretrained("facebook/dinov2-base")
```

```python
from PIL import Image
from tqdm import tqdm

# Define the number of images to process (half of the images)
num_images_to_process = len(all_image_paths) // 2

def extract_features(image_path):
    image = Image.open(image_path)
    inputs = image_processor(image, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.squeeze().cpu().numpy()

# Extract features for only the first half of the images
image_features = [extract_features(image_path) for image_path in tqdm(all_image_paths[:num_images_to_process])]
```

```python
import numpy as np
```

```python
import faiss

# Convert list of features to a matrix
feature_matrix = np.vstack(image_features)

# Build the FAISS index
index = faiss.IndexFlatL2(feature_matrix.shape[1])
index.add(feature_matrix)
```

```python
import os
from PIL import Image
from tqdm import tqdm

# Function to retrieve and save images
def retrieve_and_save_images(query_image_path, top_k=10, output_folder="retrieved_images"):
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Extract features from the query image
    query_features = extract_features(query_image_path)

    # Search for similar images in the dataset
    distances, indices = index.search(np.expand_dims(query_features, axis=0), top_k)

    # Retrieve and save the similar images
    for i in tqdm(range(top_k)):
        retrieved_image_path = all_image_paths[indices[0][i]]
        image = Image.open(retrieved_image_path)
        image.save(os.path.join(output_folder, f"retrieved_image_{i + 1}.jpg"))

# Test the retrieval and saving
query_image_path = "/content/Reference Image/query.jpg"  # Replace with the path to your query image
retrieve_and_save_images(query_image_path)
```
